# Obtener los eigenvectores de una matriz

```python
import numpy as np
```

```python
A = np.random.random((3,3)) # Generamos una matriz de 4x4 con valores aleatorios
print(A)
```

```python
w,v = np.linalg.eig(A)
```

```python
# Desplegamos los eigenvalores de la matriz A
print(w)
```

```python
# Desplegamos los eigenvectores de la matriz A
print(v)
```

```python
type(v)
```

La función *eig* de la sublibreria *linalg* de Numpy nos permite obtener los eigenvectores e eigenvectores de una matriz. Para hacerlo debemos declarar dos variables: en la primera se guardaran los valores, mientras que en la segunda los vectores.

Los eigenvectores se guardan como un np.ndarray (array de Python), donde cada **columna** corresponde a un eigenvector.

# Graficas logarítmicas

De acuerdo con la documentación de Matplotlib, hay tres formas de hacer una grafica logarítmica:
* Usando la función *semilogx()*
* Usando la función *semilog(y)*
* Usando la función *loglog()*

Veamos los ejemplos que propone.


```python
# Para hacer las gráficas necesitamos numpy y matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(0.01,20,0.01) # Generamos un conjunto de datos
```

Para un eje Y en escala logarítmica

```python
plt.figure()
plt.semilogy(t, np.exp(-t/5))
plt.title('Eje Y logarítmico')
plt.grid() # Muestra el mallado en la gráfica
plt.show()
```

Para un eje X en escala logarítmica

```python
plt.figure()
plt.semilogx(t, np.sin(2*np.pi*t))
plt.title('Eje X logarítmico')
plt.grid() # Muestra el mallado en la gráfica
plt.show()
```

Logaritmo en el eje X y en el eje Y

```python
plt.figure()
plt.loglog(t, 20*np.exp(-t/10))
plt.title('Ambos ejes logarítmicos')
plt.grid() # Muestra el mallado en la gráfica
plt.show()
```

En principio la forma para hacer gráficas logarítmicas es la misma que si hicieramos una gráfica normal. Necesitamos los datos del eje X (en este caso nuestra variable t) y los valores del eje Y. Lo único que hay que "decirle" a Matplotlib es que lo grafique de forma logarítmica.

Más información sobre todos los tipos de gráficos soportados [aquí](https://matplotlib.org/stable/tutorials/introductory/sample_plots.html#sphx-glr-tutorials-introductory-sample-plots-py)

# Múltiples gráficas en una misma imagen

Para poner más de una gráfica en un mismo espacio usaremos la función *subplot()*. Sin embargo, tenemos que definir el número de filas y de columnas que tendrá nuestro espacio y, debemos rellenar esos espacios con su gráfica correspondiente.

```python
# Generamos unos datos para poder gráficar

x = np.arange(0,100,1)
y1 = x*2
y2 = x**2
y3 = np.sin(x) # Seno de x
y4 = np.cos(x) # Coseno de x
```

En la celda anterior tenemos un espacio de 100 muestras que van desde 0 hasta 100. Después generamos cuatro conjuntos de datos, con ellos haremos cuatro gráficas en una misma figura.

Hay dos formas posibles para hacer las gráficas, en este caso veremos la forma que encuentro más sencilla.

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4,1) # Un espacio de 4 filas y una columna
plt.suptitle('Muchas gráficas')

ax1.plot(x,y1,'o-')
ax2.plot(x,y2,'y')
ax3.plot(x,y3,'b')
ax4.plot(x,y4,'r--')

plt.show()
```

Más información [aquí](https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplot.html)

Lo que hicimos en la celda anterior fue generar una figura y cuatro subespacios a graficar. Despues hicimos la gráfica en específico para cada subespacio. Siend *ax1* nuestra primera gráfica y así sucesivamente.

La "forma" que tendrá nuestra figura principal se define al inicio.

Se pueden poner las gráficas varias columnas.

```python
# Un espacio de una fila y cuatro columnas
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1,4)
plt.suptitle('Muchas gráficas')

ax1.plot(x,y1,'o-')
ax2.plot(x,y2,'y')
ax3.plot(x,y3,'b')
ax4.plot(x,y4,'r--')

plt.show()
```

Se puede hacer pero el resultado no es tan estético que digamos

```python
# Un espacio de una fila y cuatro columnas
fig, ax = plt.subplots(2,2, sharex=True)
plt.suptitle('Muchas gráficas')

ax[0,0].plot(y1)
ax[0,1].plot(y2)
ax[1,0].plot(y3)
ax[1,1].plot(y4)


plt.show()
```

Este ejemplo lo tomé de [aquí](https://interactivechaos.com/es/manual/tutorial-de-matplotlib/la-funcion-subplots)

En la celda anterior vimos la forma de gráficar en varias columnas y renglones. El procedimiento es distinto ya que solo declaramos una variable *ax* y después (usando indexación) pusimos las gráficas exactamente donde queríamos que estuvieran.

Por último, "hicimos" algo de trampa al indicar que el eje X será la misma en todas las gráficas con la instrucción *sharex=True*, sin embargo, este valor se pone por defecto en **False**.

# Para hacer las gráficas más grandes

Para hacer que las gráficas sean de un mayor tamaño sólo deben poner esta línea:

plt.rcParams['figure.figsize'] = (16,9)

después de haber escrito la líneal *matplotlib.pyplot*.

El valor dentro del paréntesis es lo que cambiará el tamaño.

Un tamaño de 16,9 hará ver a las gráficas como si fueran una pantalla plana HD.

Un tamaño de 4,3 dara un aspecto de una televisión cuadrada.

En un tamaño de 21,9 la gráfica se vera como si estuviera en un monitor *widescreen*.

```python
x = np.arange(0,50)
y = x**(0.5)
```

```python
plt.rcParams['figure.figsize'] = (16,9)

plt.plot(x,y)
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (4,3)

plt.plot(x,y)
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (21,9)

plt.plot(x,y)
plt.show()
```

# Poner un color para cada barra en una gráfica de barras

Para poner colores a cada una de las gráficas se debe generar una lista previa que contenga los nombres de dichos colores.

Para usar estos colores deben escribir la palabra *color* al momento que están generando las gráficas.

Para una lista de todos los colores disponibles en matplotlib pueden ir [aquí](https://matplotlib.org/stable/gallery/color/named_colors.html).

```python
plt.rcParams['figure.figsize'] = (10,10) # Tamaño de la gráfica

color_list = ['yellow','green','aqua','skyblue'] # Nombre de los colores en formato CCS

divisiones = ['Amarillo','Verde','Aqua','Azul cielo'] # Grupos
valores = [100,120,200,250] # Ocurrencias

plt.figure()
plt.bar(divisiones,valores,color=color_list)
plt.title('Diferentes colores')
plt.show()
```

El proceso es igual a hacer una gráfica de barras normal, con el agregado de que ahora indicamos los colores a usar.

# Poner los porcentajes en una gráfica de pastel



```python
# Generamos los datos

labels = ['Item 1','Item 2','Item 3','Item 4','Item 5']
data = [12, 32, 25, 28, 43]

# Haciendo la gráfica de pastel
plt.figure()
plt.pie(data,labels=labels, autopct='%1.1f%%')
plt.title('Gráfica con porcentajes')
plt.legend(title='Items')
plt.show()
```

```python
# Generamos los datos

labels = ['Item 1','Item 2','Item 3','Item 4','Item 5']
data = [12, 32, 25, 28, 43]

# Haciendo la gráfica de pastel
plt.figure()
plt.pie(data,labels=labels, autopct='%5.2f%%')
plt.title('Gráfica con porcentajes')
plt.legend(title='Items')
plt.show()
```

El proceso es igual al de una gráfica de pastel normal. Lo único que hay que agregar es la opción *autopct*. Dependiendo el texto que escribamos será la cantidad de decimales desplegados.

* 1.2f --> Un solo decimal
* 5.2f --> Notación 'científica'

Mas información [aquí](https://matplotlib.org/stable/gallery/pie_and_polar_charts/pie_demo2.html#sphx-glr-gallery-pie-and-polar-charts-pie-demo2-py)

# Cambiar el índice default de un DataFrame en Pandas

```python
import pandas as pd
```

```python
# Generamos un dataset a partir de un diccionario

musicos = {
    'Nombre':['Bethoven','Chopin','Vivaldi','Wagner','Mozart'],
    'Nacionalidad':['Alemán','Polaco','Italiano','Alemán','Austriaco'],
    'Año de nacimiento':[1770,1810,1741,1813,1756]
}

df = pd.DataFrame(musicos)
```

```python
df.head()
```

```python
# Cambiando el index por uno personalizado
index = {'a','b','c','d','e'}

df = pd.DataFrame(musicos,index)
```

```python
df.head()
```

```python
# Otra forma
# Generamos un dataset a partir de un diccionario

musicos = {
    'Nombre':['Bethoven','Chopin','Vivaldi','Wagner','Mozart'],
    'Nacionalidad':['Alemán','Polaco','Italiano','Alemán','Austriaco'],
    'Año de nacimiento':[1770,1810,1741,1813,1756]
}

df1 = pd.DataFrame(musicos)
df1.head()
```

```python
# Usaremos la variable 'Año de nacimiento' como nuevo índice
df1.set_index(['Año de nacimiento'], inplace=True)
df1.head()
```

La operación se puede hacer pero se "desordenan" los datos del DataFrame, por lo que en la mayoría de los casos no es algo deseado. Pero, si su problema lo requiere se puede realizar.
